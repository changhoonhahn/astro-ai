# Example 0
Let's return to the example from a few weeks ago

> Imagine a hypothetical simplistic example, where we point our telescope to observe the light from a single star with true flux $F_{\rm true}$, which remains constant with time. We will ignore any systematic errors. We perform $N$ measurements with our telescope, where the $i^{\rm th}$ measurement report an observed flux $F_i$ with uncertainty/error $e_i$.

Lets generate a synthetic observation

```python
F_true = 1000
F_obs = np.random.poisson(lam=F_true)
```

Let's write down the a simulation of the data generation process. In this case this is trivial because we know exactly the data generation process

```python
def simulation(F_true): 
    return np.random.poisson(lam=F_true)
```

First, we want to sample $(theta', \mathcal{D}') \sim p(\theta, \mathcal{D})$


As in ABC, this involves repeating two steps
1. sample the prior: $\theta'\sim p(\theta)$
2. run simulation: $\mathcal{D}' \sim F(\theta')$

For our prior, lets use
$$p(\theta) = \mathcal{N}(1002, 4)$$

```python
def sample_prior(): 
    # sample prior
    return 1002. + 4. * np.random.normal()
```

```python
Nsample = 10000

samples = np.zeros((Nsample, 2))
for i in range(Nsample): 
    theta_p = sample_prior()
    data_p = simulation(theta_p)
    samples[i,:] = np.array([theta_p, data_p])
samples = jnp.array(samples)
```

```python
plt.scatter(samples[:,0], samples[:,1], s=1)
plt.xlabel(r'$\theta$', fontsize=25)
plt.ylabel(r'$\mathcal{D}$', fontsize=25)
```

Now lets use variational inference to estimate this joint distribution. First, we need to select our surrogate distribution $q_\phi(\theta, \mathcal{D})$. 

We'll use a simple 2D Gaussian, so $\phi$ is the mean and covariance matrix

```python
from jax.scipy.linalg import cho_factor, cho_solve

def log_q(x, mean, cov):
    """
    x: (..., 2)
    mean: (2,)
    cov: (2,2) positive-definite covariance matrix
    """
    # Cholesky factorization
    L = jnp.linalg.cholesky(cov)
    
    diff = x - mean
    
    # Solve ^{-1}(x-) using Cholesky
    sol = jnp.linalg.solve(L, diff.T).T  
    
    # log determinant: log det  = 2 * sum(log(diag(L)))
    logdet = 2.0 * jnp.sum(jnp.log(jnp.diag(L)))

    return -0.5 * (2 * jnp.log(2 * jnp.pi) + logdet + jnp.sum(sol**2, axis=-1))
```

Now lets determine $\phi$ by minimizing the KL divergence between $p(\theta, \mathcal{D})$ and $q_\phi(\theta, \mathcal{D})$: 

$$D_{\rm KL} (p\parallel q_\phi) \approx \sum\limits_{(\theta',\mathcal{D}')\sim p(\theta, \mathcal{D})} \left[ - \log q_\phi(\theta',\mathcal{D}')\right]$$

```python
class Params(NamedTuple):
    mu:  jnp.ndarray
    A: jnp.ndarray
    
def KL(x, param: Params, key):
    # KL divergence D_KL(p || q) = sum -log q(x) for x ~ q(x)
    
    cov = param.A @ param.A.T + 1e-3 * jnp.eye(2)  #ensure its positive definite

    # estimate KL
    return -jnp.mean(log_q(x, param.mu, cov))

def loss_fn(params, key):
    return KL(samples, params, key)

@jax.jit
def step(params, opt_state, key):
    # optimization step
    key, subkey = jax.random.split(key)
    loss, grads = jax.value_and_grad(loss_fn)(params, subkey)
    updates, opt_state = opt.update(grads, opt_state, params)
    params = optax.apply_updates(params, updates)
    return params, opt_state, key, loss    
```

```python
# initialize the optimizer
params = Params(mu=jnp.array([1002.0, 1002.0]), A=jnp.array([[4, 0.], [0., 30.]]))  # mu, rho
opt = optax.adam(0.01)
opt_state = opt.init(params)

# ----- Train -----
for t in range(2001):
    params, opt_state, key, loss = step(params, opt_state, key)
```

Lets check how well $q_\phi(\theta, \mathcal{D})$ approximates $p(\theta, \mathcal{D})$

```python
q_samples = np.random.multivariate_normal(np.array(params.mu), np.array(params.A @ params.A.T), size=100000)
```

```python
fig = corner.corner(np.array(samples))
fig = corner.corner(q_samples, color='C1', fig=fig)
```

looks pretty good! Now we can move onto ABC using the surrogate $q_\phi$. The advantage here is that $q_\phi$ is extremely cheap to sample, so we can generate millions of samples that pass our distance critiera

```python
q_samples = np.random.multivariate_normal(np.array(params.mu), np.array(params.A @ params.A.T), size=10000000)
close_enough = np.abs(q_samples[:,1] - F_obs) < 1
print('%i have D \approx D_obs' % np.sum(close_enough))
```

```python
plt.hist(q_samples[:,0][close_enough], range=(960, 1040), bins=40)
plt.axvline(F_true, color='k', linestyle='--')
```

This is a bit of a silly example since the simulator is very cheap to run. But in many astrophysical applications the simulator of the data generation process is very expensive (cosmological simulations), then it would make sense to sample the joint distribution using the surrogate!

# Example 1
Now that was a very simple 1D example. Lets consider a slightly more interesting example of fitting a line to some data and see how SBI with variational inferenceh olds up. First, we'll go through the example with a slight modification: Gaussian priors rather than uniform priors. 

```python
a_true, b_true = 0, 1.75
x = np.array([1, 1.5, 2, 2.5])
y = np.array([1.87640523, 2.59001572, 3.4978738 , 4.47408932])
yerr = 0.1
```

```python
plt.errorbar(x, y, yerr=yerr, fmt='.k')
```

```python
def simulation(theta): 
    a, b = theta
    noise = yerr * np.random.normal(size=len(x))
    return a + b * x + noise

def sample_prior(): 
    # sample a and b from uniform prior
    _a = -0.5 + np.random.normal() 
    _b = 5 + 3*np.random.normal()
    return [_a, _b]
```

same as above, sample $\theta', \mathcal{D}' \sim p(\theta, \mathcal{D})$

```python
Nsample = 500000

samples = np.zeros((Nsample, 6))
for i in range(Nsample): 
    theta_p = sample_prior()
    data_p = simulation(theta_p)
    samples[i,:] = np.concatenate([theta_p, data_p])
```

```python
class Params(NamedTuple):
    mu:  jnp.ndarray
    A: jnp.ndarray
    
def KL(x, param: Params, key):
    # KL divergence D_KL(p || q) = sum -log q(x) for x ~ q(x)
    
    cov = param.A @ param.A.T + 1e-3 * jnp.eye(6)  #ensure its positive definite

    # estimate KL
    return -jnp.mean(log_q(x, param.mu, cov))

def loss_fn(params, key):
    return KL(samples, params, key)

@jax.jit
def step(params, opt_state, key):
    # optimization step
    key, subkey = jax.random.split(key)
    loss, grads = jax.value_and_grad(loss_fn)(params, subkey)
    updates, opt_state = opt.update(grads, opt_state, params)
    params = optax.apply_updates(params, updates)
    return params, opt_state, key, loss    
```

minimize KL divergence to get $\phi$

```python
# initialize the optimizer
params = Params(mu=jnp.array([0., 2., 5., 5., 10., 10.]), A=jnp.diag(np.array([1, 3., 3., 4., 6, 7.5])))  # mu, rho
opt = optax.adam(0.01)
opt_state = opt.init(params)

# ----- Train -----
for t in range(1001):
    params, opt_state, key, loss = step(params, opt_state, key)
```

```python
q_samples = np.random.multivariate_normal(np.array(params.mu), np.array(params.A @ params.A.T), size=100000)
```

```python
fig = corner.corner(np.array(samples))
fig = corner.corner(q_samples, color='C1', fig=fig)
```

comparison of $q_\phi$ to $p$ looks pretty good by eye. Lets move on to the ABC approximation

```python
q_samples = np.random.multivariate_normal(np.array(params.mu), np.array(params.A @ params.A.T), size=10000000)
```

```python
rho = np.sqrt(np.sum((q_samples[:,2:] - y[None, :])**2, axis=1))
```

```python
epsilon = 0.5
print('reject %f percent of simulations' % (100.*(1-np.mean(rho < epsilon))))
```

```python
fig = corner.corner(q_samples[:,:2][rho < epsilon], labels=['a', 'b'])
corner.overplot_lines(fig, [a_true, b_true], color="C1")
corner.overplot_points(fig, np.array([a_true, b_true])[None], marker="s", color="C1")
```

# Example 1.b
Now what happens if we change the prior back to the uniform priors

```python
def simulation(theta): 
    a, b = theta
    noise = yerr * np.random.normal(size=len(x))
    return a + b * x + noise

def sample_prior(): 
    # sample a and b from uniform prior
    _a = -0.5 + np.random.uniform() 
    _b = 10*np.random.uniform()
    return [_a, _b]

Nsample = 500000

samples = np.zeros((Nsample, 6))
for i in range(Nsample): 
    theta_p = sample_prior()
    data_p = simulation(theta_p)
    samples[i,:] = np.concatenate([theta_p, data_p])
```

```python
class Params(NamedTuple):
    mu:  jnp.ndarray
    A: jnp.ndarray
    
def KL(x, param: Params, key):
    # KL divergence D_KL(p || q) = sum -log q(x) for x ~ q(x)
    
    cov = param.A @ param.A.T + 1e-3 * jnp.eye(6)  #ensure its positive definite

    # estimate KL
    return -jnp.mean(log_q(x, param.mu, cov))

def loss_fn(params, key):
    return KL(samples, params, key)

@jax.jit
def step(params, opt_state, key):
    # optimization step
    key, subkey = jax.random.split(key)
    loss, grads = jax.value_and_grad(loss_fn)(params, subkey)
    updates, opt_state = opt.update(grads, opt_state, params)
    params = optax.apply_updates(params, updates)
    return params, opt_state, key, loss    

# initialize the optimizer
params = Params(mu=jnp.array([0., 2., 5., 5., 10., 10.]), A=jnp.diag(np.array([1, 3., 3., 4., 6, 7.5])))  # mu, rho
opt = optax.adam(0.01)
opt_state = opt.init(params)

# ----- Train -----
for t in range(1001):
    params, opt_state, key, loss = step(params, opt_state, key)    
```

```python
q_samples = np.random.multivariate_normal(np.array(params.mu), np.array(params.A @ params.A.T), size=100000)
fig = corner.corner(np.array(samples))
fig = corner.corner(q_samples, color='C1', fig=fig)
```

Now you can see why I chose Gaussian priors earlier. With uniform priors, we're confronted with the same issue as variational inference. Our multivariate Gaussian surrogate cannot estimate $p(\theta, \mathcal{D})$ accurately. In this case, there's no way to accurately estimate the posterior with $q_\phi$.
